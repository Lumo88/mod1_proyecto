```python
import random # Activamos la funci√≥n 'aleatoria' para que la m√°quina pueda jugar

num_pregunta = 1 # Variables iniciales cada vez que comienza el juego
puntuacion_concursante = 0

preguntas_opciones = [{"pregunta" : "¬øcual es el planeta mas cercano al sol?", "opcion a" : "venus", "opcion b" : "mercurio", "opcion c" : "marte", "respuesta" : "b"}, 
{"pregunta" : "¬øcual es el metal cuyo simbolo quimico es 'fe'?", "opcion a" : "hierro", "opcion b" : "plata", "opcion c" : "cobre", "respuesta" : "a"}, 
{"pregunta" : "¬øcual es el rio mas largo del mundo?", "opcion a" : "amazonas", "opcion b" : "nilo", "opcion c" : "yangtse", "respuesta" : "a"}, 
{"pregunta" : "¬øcuantos dias tiene un a√±o bisiesto?", "opcion a" : 364, "opcion b" : 365, "opcion c" : 366, "respuesta" : "c"}, 
{"pregunta" : "¬øque gas necesitan las plantas para realizar la fotosintesis?", "opcion a" : "dioxido de carbono", "opcion b" : "oxigeno", "opcion c" : "nitrogeno", "respuesta" : "a"},
{"pregunta" : "¬øquien pinto 'la gioconda'?", "opcion a" : "miguel angel", "opcion b" : "leonardo da vinci", "opcion c" : "picasso", "respuesta" : "b"},
{"pregunta" : "¬øque filosofo dijo 'solo se que no se nada'?", "opcion a" : "platon", "opcion b" : "artistoteles", "opcion c" : "socrates", "respuesta" : "c"},
{"pregunta" : "¬øque parte del cuerpo humano no crece desde que nacemos?", "opcion a" : "huesecillos del oido", "opcion b" : "ojos", "opcion c" : "cartilago de la nariz", "respuesta" : "b"},
{"pregunta" : "¬øcomo se llama el dinero que utiliza un pais como medio de pago?", "opcion a" : "capital", "opcion b" : "ahorro", "opcion c" : "moneda", "respuesta" : "c"},
{"pregunta" : "¬øen que a√±o comenzo la segunda guerra mundial?", "opcion a" : 1914, "opcion b" : 1945, "opcion c" : 1939, "respuesta" : "c"}
] # Fijamos una lista de diccionarios con las preguntas, opciones y respuestas correctas 
preguntas_usadas = preguntas_opciones.copy() # Creamos una copia de la lista de preguntas para meter ah√≠ las preguntas utilizadas durante el concurso

print("¬°¬°BIENVENIDOS AL ESPERADO CONCURSO DE CULTURA GENERAL!!")
print("------------------------------------")
print("A continuaci√≥n,responder√°s a 5 preguntas con 3 opciones de respuestas. Debes responder con una letra: 'a', 'b', o 'c'") ###
print("************************************")

while num_pregunta <= 5: # Bucle princicpal del juego: 5 preguntas a responder 
    
    confirmacion = ""

    while confirmacion not in ["si", "no"]: # Bucle de confrimaci√≥n para continuar el concurso o no
        confirmacion = input("Quieres continuar concursando? (si/no)").lower() # El concursante elige una respuesta en concreto

        if confirmacion == "si": # El concurso continua
            print("¬°Continuamos entonces!")

        elif confirmacion == "no": # El concurso se para
            print("Saliendo del concurso...")
            break ##### NO FUNCIONA

        else: # Volver a poner bien la palabra descrita
            print("‚ö†Ô∏è Opci√≥n no v√°lida ‚ö†Ô∏è Int√©ntalo de nuevo")

    if num_pregunta == 5: # T√≠tulo de las preguntas
        print("------------- PREGUNTA FINAL -------------")
    
    else:
        print(f"------------- PREGUNTA {num_pregunta} -------------")

    pregunta_actual = random.choice(preguntas_usadas) # La m√°quina elige una pregunta de manera aleatoria de la lista copiada
    indice = preguntas_usadas.index(pregunta_actual) # Buscamos en qu√© posici√≥n (el √≠ndice) est√° la pregunta utilizada
    preguntas_usadas.pop(indice) # Eliminamos la pregunta que se ha utilizado de la lista copiada para que no se repita
    print(f"Pregunta: {pregunta_actual["pregunta"]}")
    print(f"Opcion a: {pregunta_actual["opcion a"]}")
    print(f"Opcion b: {pregunta_actual["opcion b"]}")
    print(f"Opcion c: {pregunta_actual["opcion c"]}")

    print("------------------------------------")

    respuesta_concursante = ""

    while respuesta_concursante not in ["a", "b", "c"]: # En el caso de que el jugador no escriba correctamente alguna letra descrita, tendr√° que volverlo a escribir hasta que la escriba bien
        respuesta_concursante = input("¬øCu√°l es tu respuesta? ¬øLa opci√≥n 'a', 'b' o la 'c'?").lower() # El concursante elige una respuesta en concreto
        
        if respuesta_concursante != ["a", "b", "c"]:
            print(f"‚ö†Ô∏è Opci√≥n no v√°lida ‚ö†Ô∏è Int√©ntalo de nuevo")
            continue # NO FUNCIONA

    if respuesta_concursante == pregunta_actual["respuesta"]: # Condicional clave: La respuesta es correcta o incorrecta
        puntuacion_concursante += 1 # Se suma 1 punto en el caso de qu el arespuesta sea correcta 
        print(f"Tu respuesta es la opcion: {respuesta_concursante}")
        print(f"‚úÖ ¬°Enhorabuena! ‚úÖ ¬°Has acertado!")
        print(f"Tu puntuaci√≥n actual es de: {puntuacion_concursante} punto/s")
    
    else:
        print(f"Tu respuesta es la opcion: {respuesta_concursante}")
        print(f"‚ùå ¬°Incorrecto! ‚ùå La respuesta correcta es la opci√≥n: {pregunta_actual['respuesta']}")
    
    num_pregunta += 1 # Contador de preguntas

print("************************************")
print("üèÜ ¬°FIN DEL CONCURSO! üèÜ")
print (f"Tu puntuaci√≥n final es de: {puntuacion_concursante} punto/s")
print("------------------------------------")
print("¬°Gracias por participar! Te estaremos esperando para la pr√≥xima ü§ó")                                   
```
